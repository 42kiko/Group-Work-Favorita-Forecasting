```python
import plotly.express as px

df_meat_sorted = df_meat.sort_values("date")

fig = px.line(
    df_meat_sorted,
    x="date",
    y="unit_sales",
    title="Unit Sales over Time for Item 1976284",
)

fig.update_layout(xaxis_title="Date", yaxis_title="Unit Sales")

fig.show()
```

```python
df_transactions["date"] = pd.to_datetime(df_transactions["date"])

df_2016 = df_transactions[
    (df_transactions["store_nbr"] == 39)
    & (df_transactions["date"] >= "2016-01-01")
    & (df_transactions["date"] < "2017-01-01")
]

daily_transactions = df_2016.groupby("date")["transactions"].sum().reset_index()

fig = px.line(
    daily_transactions,
    x="date",
    y="transactions",
    title="Daily Transactions Value - Store 39 (2016)",
)

fig.update_layout(xaxis_title="Date", yaxis_title="Transactions")

fig.show()
```

```python
df_train["date"] = pd.to_datetime(df_train["date"])

df_2016 = df_train[
    (df_train["store_nbr"] == 39)
    & (df_train["date"] >= "2016-01-01")
    & (df_train["date"] < "2017-01-01")
]

daily_units = df_2016.groupby("date")["unit_sales"].sum().reset_index()

fig = px.line(
    daily_units, x="date", y="unit_sales", title="Daily Units Sold - Store 39 (2016)"
)
fig.update_layout(xaxis_title="Date", yaxis_title="Units Sold")
fig.show()
```
